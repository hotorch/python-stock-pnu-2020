### 배경
모멘텀 현상은 행동재무학에서 일컫는 군집 행동, 정박 효과, 확증 편향, 처분 효과 등의 행동 편향에 의해 발생한다고 알려져 있다.  
- 군집 행동 : 다수 그룹의 행동을 따라하는 경향  
- 정박 효과 : 정보를 처음 제공받은 시점에 지나치게 의존하는 경향  
- 확증 편향 : 본인의 믿음과 반대되는 정보를 무시하는 경향  
- 처분 효과 : 수익이 난 주식을 금방 팔고, 손해 봅 주식을 계속 보유하는 경향  

위의 4개 편향들 살펴보면 다들 각자 위의 경험이 다들 있을 것일텐데, 이를 활용해서 돈을 잘먹고 나오는게 중요하다고 생각합니다.  

### 간단 개념 
- 모멘텀
: "특정 날짜 전"의 주식 가격에 비해 "현재 날짜"의 주식 가격이 얼마나 증가 혹은 감소했는지 알려주는 지표  


- ex) 어제보다 오늘 코스피가 2% 상승 ->, "아, 코스피 하루 모멘텀이 2%네"  
1년전에 비해 20%가 올랐다면 "코스피 1년 모멘텀이 20%"  


- 상대 모멘텀  
특정 기간 동안 상대적으로 수익률이 좋았던 n개 종목을 구함  


- 절대 모멘텀  
자산의 가치가 상승하고 있을 때만 투자하고, 그렇지 않을 떄는 단기 국채를 매수하거나 현금을 보유하는 전략  


- 모멘텀 전략?
현재주가와 과거 주가를 비교해서 현재주가가 높으면 매수, 낮으면 매도  


- 약점  
횡보장에서 구조적으로 손실이 누적됨. 일시적인 급락과 급등이 연이어 이어지는 시장이면서, 큰 추세없이 변동성만 큰 시장이 되면 모멘텀 전략이 잘 작동하지 않는다고 함

묶은(groupby)결과를 list와 dict로 변환하면 개별 섹터(Sector)별로 종목리스트를 구분할 수 있습니다.

```python
sectors = dict(list(nasdaq.groupby('Industry')))

print('count:', len(sectors))
list(sectors.keys())[:10]
```

```python
it_sec = sectors['소프트웨어 및 IT서비스']

print('row count:', len(it_sec))
it_sec.head(10)
```

## 개별 종목의 가격 데이터 가져오기
302개 다들고 오는게 조금 많으니까 적당히 들고와보겠습니다.

```python
it_sec = it_sec.iloc[0:20]
```

```python
it = pd.DataFrame()

for ix, row in it_sec.iterrows():
    code, name = row['Symbol'], row['Name']
    print(code, name)
    # 개별 종목의 가격을 가져옵니다
    df = fdr.DataReader(code, '2020-01-01', '2020-11-07')

    # 가격 데이터의 종가(Close)를 컬럼으로 추가합니다
    # (컬럼명은 종목명을 지정합니다)
    it[name] = df['Close']
```

```python
# 결측치 있는 column 잠시 지우겠습니다
it = it.dropna(axis=1)
print(it.shape)
it
```

## 수익률 계산하기

```python
acc_rets = it / it.iloc[0] - 1.0
acc_rets
```

수익률 반환

```python
returns = acc_rets.iloc[-1]
returns
```

수익률 내림차순

```python
returns.sort_values(ascending=False) # zoom 수익률이 627%
```

```python
returns.mean() # 기술주 평균 수익률 올해(10개월) 87%
```

## 다양한 기간에 대한 수익률
기간을 달리해서 수익률 추출

```python
df = it['2020-10-01':'2020-10-31'] # 특정 기간(12월 1달) 동안
acc_rets = df / df.iloc[0] - 1.0
acc_rets.iloc[-1]
```

'2020-10-31' 시점을 기준으로 과거 5일, 10일, 20일, 60일, 120일, 240일 각각 수익률을 구해봅니다.

각 기간의 (시작날짜, 끝날짜)는 다음과 같이 생성할 수 있습니다.

```python
from datetime import datetime, timedelta

the_day = datetime(2020, 10, 31)

for days in [5, 10, 20, 60, 120, 180, 240]:
    start = the_day - timedelta(days)
    end = the_day
    print(start, '~', end)
```

```python
from datetime import datetime, timedelta

the_day = datetime(2020, 10, 31)
row_dict = {}
for days in [5, 10, 20, 60, 120, 180, 240]:
    start = the_day - timedelta(days)
    end = the_day

    df = it[start:end] # 특정 기간
    acc_rets = df / df.iloc[0] - 1.0
    row_dict[days] = acc_rets.iloc[-1] 
```

## 데이터프레임 Style 부여

```python
def color_negative_red(val):
    color = 'red' if val < 0 else 'black'
    return 'color: %s' % color

def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

df_rets.style.\
    applymap(color_negative_red).\
    apply(highlight_max)
```
